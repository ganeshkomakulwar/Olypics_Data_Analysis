```python
ath_data = pd.read_csv('athlete_events.csv')
noc = pd.read_csv('noc_regions.csv')
```

```python
#Merging Athlete Data And noc Data
```

```python
merged = pd.merge(ath_data,noc,on='NOC',how='left')
```

```python
merged.shape
```

```python
merged.head(10)
```

```python
merged.describe()
```

```python
#Removing Notes Column From The Merged Data
```

```python
merged.drop(columns = ['notes'],inplace=True)
```

```python
merged.head(10)
```

```python
#Distribution Of Gold Medals
```

```python
merged['Medal'].value_counts()
```

```python
golddf = merged[merged['Medal']=='Gold']
golddf.head(10)
```

```python
#Distribution Of Silver Medals
```

```python
silverdf = merged[merged['Medal']=='Silver']
silverdf.head(10)
```

```python
#Distribution Of Bronze Medals
```

```python
Bronzedf = merged[merged['Medal']=='Bronze']
Bronzedf.head(10)
```

```python
#Missing Value Check In Gold Data
```

```python
golddf.isnull().sum()
```

```python
golddf = golddf[np.isfinite(golddf['Age'])]
golddf = golddf[np.isfinite(golddf['Height'])]
golddf = golddf[np.isfinite(golddf['Weight'])]
golddf.isnull().sum()
```

```python
#Missing Value Check In Silver Data
```

```python
silverdf.isnull().sum()
```

```python
silverdf = silverdf[np.isfinite(silverdf['Age'])]
silverdf = silverdf[np.isfinite(silverdf['Height'])]
silverdf = silverdf[np.isfinite(silverdf['Weight'])]
silverdf.isnull().sum()
```

```python
#Missing Value Check In Bronze Medals
```

```python
Bronzedf.isnull().sum()
```

```python
Bronzedf = Bronzedf[np.isfinite(Bronzedf['Age'])]
Bronzedf = Bronzedf[np.isfinite(Bronzedf['Height'])]
Bronzedf = Bronzedf[np.isfinite(Bronzedf['Weight'])]
Bronzedf.isnull().sum()
```

```python
#Ploting The Distribution
```

```python
plt.figure(figsize=(20,10))
plt.tight_layout()
sns.countplot(x=golddf['Age'])
plt.title('Age Distribution On Gold Medals')
```

```python
plt.figure(figsize=(20,10))
plt.tight_layout()
sns.countplot(x=silverdf['Age'])
plt.title('Age Distribution On Silver Medals')
```

```python
plt.figure(figsize=(20,10))
plt.tight_layout()
sns.countplot(x=Bronzedf['Age'])
plt.title('Age Distribution On Bronze Medal')
```

```python
#Finding In Which Sport Athlete Get Gold Medal Having Age Greater Than 50 Years
```

```python
golddf[golddf['Age']>=50]['Sport']
plt.figure(figsize=(20,10))
plt.tight_layout()
sns.countplot(x=golddf[golddf['Age']>=50]['Sport'])
plt.title('Athletes Won Gold Medal In Sports Having Age Greater Than 50 Years')
```
